## Fig 3e: NAD amount

```python
folder = "labelling_experiments_cell_lines"
folder = os.path.join(infolder_base, folder)
fig_prefix = "fig3d_"
```

```python
glob_list = {
    "293_?.?": "293T",
    "CytoPARP_?.?": "cytoPARP",
    "ER_PARP_?.?": "erPARP",
    "mitoPARP_?.?": "mitoPARP",
    "pexPARP_?.?": "pexPARP",
}

dfs = []
for g in glob_list:
    full_glob = os.path.join(folder, g + suffix)
    dfs.append(analyse_rawfiles(full_glob))
df = pd.concat(dfs)
```

```python
# total intracellular NAD concentrations and SDs in nmol/mg protein
nad_meas = pd.read_excel(
    "../../data/NAD_conc_MS.xlsx",
    sheet_name="Summary",
    nrows=10,
    skiprows=3,
    usecols=[1, 2, 3],
    index_col=0,
)
nad_meas = nad_meas.rename(
    {
        293: "293_?.?",
        "mP": "mitoPARP_?.?",
        "cP": "CytoPARP_?.?",
        "pP": "pexPARP_?.?",
        "ERP": "ER_PARP_?.?",
    }
)

colors_nad_amount = colors_cell_lines.rename(
    {
        "293": "293_?.?",
        "mitoPARP": "mitoPARP_?.?",
        "pexPARP": "pexPARP_?.?",
        "cytoPARP": "CytoPARP_?.?",
        "erPARP": "ER_PARP_?.?",
    }
)

xlim = (-0.1, 6)
ylim = (0, 12)

# Filter out all rows with 3-AB
nad_meas = nad_meas[~nad_meas.index.str.contains(pat="3-AB")]
nad_protein_mean = nad_meas["mean (nmol/mg protein)"]
nad_protein_sd = nad_meas["SD"]
```

```python
outfile_amount = os.path.join(outfolder, f"{fig_prefix}nad_amount{outsuffix}")
outfile_amount_d = os.path.join(outfolder, f"{fig_prefix}nad_amount_detail{outsuffix}")
outfile_per = os.path.join(outfolder, f"{fig_prefix}nad_percent{outsuffix}")
outfile_per_d = os.path.join(outfolder, f"{fig_prefix}nad_percent_detail{outsuffix}")

plot_nad_amount(
    df,
    nad_mean=nad_protein_mean,
    nad_std=nad_protein_sd,
    multiplot=colors_nad_amount,
    outfile=outfile_amount,
)
plot_nad_amount(
    df,
    nad_mean=nad_protein_mean,
    nad_std=nad_protein_sd,
    multiplot=colors_nad_amount,
    outfile=outfile_amount_d,
    xlim=xlim,
    ylim=ylim,
)
plot_syn_per(df, multiplot=colors_nad_amount, outfile=outfile_per)
plot_syn_per(df, multiplot=colors_nad_amount, xlim=ylim, outfile=outfile_per_d)
```

## Fig 4b: 3AB Simulations for percentage change

```python
folder = "3ABsimulations"
folder = os.path.join(infolder_simu_base, folder)
fig_prefix = "fig4b_"

experiments = {
    "293": ["Fluxes_293_3AB_stable.txt", "Fluxes_293_3AB_washout.txt"],
    "pexPARP": ["Fluxes_pexPARP_stable_3AB.txt", "Fluxes_pexPARP_washout.txt",],
}
```

## Fig 4c: 3AB Simulations for NAD amount

```python
folder = "3ABsimulations"
folder = os.path.join(infolder_simu_base, folder)
fig_prefix = "fig4c_"

file_list = [
    "Fluxes_293_3AB_stable.txt",
    "Fluxes_293_3AB_washout.txt",
    "Fluxes_pexPARP_stable_3AB.txt",
    "Fluxes_pexPARP_washout.txt",
]
file_list = [os.path.join(folder, f) for f in file_list]

colors = {}
colors["Fluxes_293_3AB_stable"] = "black"
colors["Fluxes_293_3AB_washout"] = "grey"
colors["Fluxes_pexPARP_stable_3AB"] = "#3399FF"
colors["Fluxes_pexPARP_washout"] = "#0000CC"
```

```python
df = read_model_output(file_list)
df["Values[Biosynthesis fluxes]"] = df["Values[Biosyntehsi fluxes]"]

for exp, data in df.groupby("Exp"):
    title = " ".join(os.path.splitext(exp)[0].split("_")[1:])
    title = "Simulation " + title
    exp_name = "_".join(os.path.splitext(exp)[0].split("_")[1:])
    outfile = os.path.join(outfolder, f"{fig_prefix}{exp_name}{outsuffix_simu}")

    plot_simu_abs(
        data, title=title, half_life=True, color=colors[exp], outfile=outfile,
    )
```
